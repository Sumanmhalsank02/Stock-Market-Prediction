```python
# Making a new dataframe with the column as the annualised returns of 
# the portfolio(closing prices of 30 stocks) 
#Calculate average annual percentage return and volatilities over a theoretical one year period

returns = Portfolio.pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ["Returns"]


returns["Volatility"] = Portfolio.pct_change().std() * np.sqrt(252)
```

```python
returns
```

```python
from sklearn.cluster import KMeans


#Format the dataframe into numpy arrays to feed in the kmeans model
data = np.asarray([np.asarray(returns["Returns"]), np.asarray(returns["Volatility"])]).T


X = data
distortions = []

for k in range(2,30):
    k_means = KMeans(n_clusters = k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)
    
plt.figure(figsize= (10,7))
plt.plot(range(2,30), distortions)
plt.grid(True)
plt.title("Elbow Curve")
plt.show()
```

```python
"""So we can sort of see that once the number of clusters reaches 5 (on the bottom axis), 
the reduction in the SSE begins to slow down for each increase in cluster number
Hence  the optimal number of clusters lies around the 5 mark """
```

```python
from scipy.cluster.vq import kmeans,vq
import seaborn as sns

#drop the relevant stock from our data
#returns.drop('DHFL',inplace=True)   

#recreate data to feed into the algorithm
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T

# computing K-Means with K = 5 (5 clusters)
centroids,_ = kmeans(data,5)

# assign each sample to a cluster
idx,_ = vq(data,centroids)

details = [(name,cluster) for name, cluster in zip(returns.index,idx)]

Cluster1 = []
Cluster2 = []
Cluster3 = []
Cluster4 = []
Cluster5 = []

for i in range(len(details)):
    if details[i][1]==0:
        Cluster1.append(details[i][0])
        
    if details[i][1]==1:
        Cluster2.append(details[i][0])
        
    if details[i][1]==2:
        Cluster3.append(details[i][0])
        
    if details[i][1]==3:
        Cluster4.append(details[i][0])
        
    if details[i][1]==4:
        Cluster5.append(details[i][0])
        
Clusters = ["Cluster1","Cluster2","Cluster3","Cluster4","Cluster5"]

labels = ["VOLTAS", "TATAPOWER", "TATACHEM", "SUNTV", "RBLBANK", "NBCC", "DHFL", "IDBI", "JUBLFOOD", "GODREJIND",
         "ASHOKA", "BAJAJELEC", "FORTIS", "JETAIRWAYS", "SONATSOFTW", "NCC", "PVR", "RAYMOND", "CENTURYPLY", "VIPIND",
         "ADANIPORTS", "ASIANPAINT", "BAJFINANCE", "CIPLA", "ITC", "LT", "MARUTI", "RELIANCE", "TCS", "TITAN"]



# some plotting using numpy's logical indexing
plt.figure(figsize=(10,7))
plt.plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om')
    
plt.plot(centroids[:,0],centroids[:,1],'sg',markersize=10, label = labels )

plt.legend(Clusters)
plt.show()

```
