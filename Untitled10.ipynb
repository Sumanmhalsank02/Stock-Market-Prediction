```python
# Adding a column Call based on the multiple conditions given
# Try using a for loop also
# for the below code ,took reference from stack overflow

def Callf(data1):
    
    if data1["Close Price"] < data1["Bollinger Low"]:
        return "Buy"
    
    elif data1["Bollinger Low"] < data1["Close Price"] < data1["Bollinger Middle"]:
        return "Hold Buy/Liquidate Short"
    
    elif data1["Bollinger Middle"] < data1["Close Price"] < data1["Bollinger High"]:
        return "Hold Short/Liquidate Buy"
    
    elif data1["Close Price"] > data1["Bollinger High"]:
        return "Short"
    
data1["Calls"] = data1.apply(Callf, axis=1)

```

```python
# importing the libraries for different Classification ALgos
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB


# Defining the INput and OUtput values
X = pd.concat([data1[["Close Price", "Bollinger High", "Bollinger Low", "Bollinger Middle"]]])
y = pd.concat([data1[["Calls"]]])
y = np.ravel(y)       # if we don't use this we get a lot of warnings


# Splitting data into testing and training part
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state =0)

#Choosing models
# LogisticRegression
lr = LogisticRegression().fit(X_train,y_train)
prediction = lr.predict(X_test)
#plt.plot(prediction)
#plt.grid(True)
print("Test Accuracy of Logistic Regression is: {}".format(lr.score(X_test,y_test)*100))


# Support Vector MAchine
svm = SVC(random_state=1).fit(X_train,y_train)
print("Test Accuracy of SVM is: {} ".format(svm.score(X_test,y_test)*100))



# Decision Tree
dtc = DecisionTreeClassifier().fit(X_train,y_train)
print("Test Accuracy for Decision Tree is: {}".format(dtc.score(X_test,y_test)*100))



# Random Forest
rt =RandomForestClassifier(n_estimators=1000, random_state=1).fit(X_train,y_train)
print("Test Accuracy for Random Forest is: {}".format(rt.score(X_test,y_test)*100))



# Naive Bayes
nb = GaussianNB().fit(X_train,y_train)
print("Test Accuracy of Naive Bayes is: {}".format(nb.score(X_test,y_test)*100))



# KNN
list = []
for i in range(1,385,1):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    prediction = knn.predict(X_test)
    list.append(knn.score(X_test,y_test)*100)
    
print("Test Accuracy of KNN is: {}".format(max(list)))

```

```python
# Classification to make a Trade Call

from sklearn.ensemble import RandomForestClassifier

Axis_df = pd.read_csv(r"C:\Users\Suman\Desktop\AICTE Internship\Large_Cap\AXISBANK.csv")



#df['pct_change'] = df.groupby(['Name'])['Amount'].pct_change()

```

```python
OC = pd.concat([Axis_df[["Open Price", "Close Price"]]])
Axis_df["PCT_OC"] = OC.pct_change(axis = 1)["Close Price"]

#OC.pct_change(axis = 1)   
# Check through the formula also
# (Close Price - OPen Price) /Open Price


HL = pd.concat([Axis_df[["Low Price", "High Price"]]])
Axis_df["PCT_HL"] = HL.pct_change(axis=1)["High Price"]
```

```python
Axis_df
```

```python
# Adding the columns for rolling mean& std of the day to day percentage of the Close Price

Axis_df["RollingMean_CP"] = Axis_df["Close Price"].pct_change().rolling(window=5).mean()
Axis_df["RollingStd_CP"] = Axis_df["Close Price"].pct_change().rolling(window=5).std()
Axis_df = Axis_df.dropna()
```

```python
# Adding the axis column based on multiple if else conditions

def Act(Axis_df):
    
    if Axis_df["Close Price"] > Axis_df["Prev Close"]:
        return 1
    
    elif Axis_df["Close Price"] < Axis_df["Prev Close"]:
        return -1
     
    
Axis_df["Action"] = Axis_df.apply(Act, axis=1)
```

```python
Axis_df.tail(336)     # for the last two years the data starts from  index 159
```

```python
# Random Forest Classification to make a Trading algorithm

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


#Defining Input and Output Values
X = pd.concat([Axis_df[["PCT_OC", "PCT_HL", "RollingMean_CP", "RollingStd_CP"]].tail(336)])
y = pd.concat([Axis_df[["Action"]].tail(336)])
y = np.ravel(y)

#splitting the data into training and testing 
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Fitting the training data and checking the accuracy
Rt = RandomForestClassifier(n_estimators=1000, random_state=1).fit(X_train,y_train)

print("Test Accuracy of the Algorithm is: {}".format(Rt.score(X_test,y_test)*100))

prediction = Rt.predict(X_test)
print()
print("prediction:\n",prediction )
```
