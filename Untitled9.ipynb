```python
# TRYING LINEAR MODEL FOR THE 'PRED' COLUMN OF THE DATAFRAME 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split     #when tried without splitting ,overfitting is facedbn
import seaborn as sns
from sklearn import metrics,linear_model
import sklearn
import math

data = pd.read_csv(r'C:\Users\Suman\Desktop\AICTE Internship\Gold.csv')

data['Change %'] = data['Change %'].str.replace(r'%', '').astype(float)  #removed % from the column values 

#X = pd.concat([data[['Open', 'High','Low','Change %']]])  #Double brackets after data
X = pd.concat([data[['Open', 'High','Low','Change %']].head(410)])
Y = pd.concat([data[['Pred']].head(410)])


#Splitting the data into training and testing set
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size= 0.4, random_state=1)
    
    
    #Create a linear Regression Object
regression_model = linear_model.Lasso(alpha=1.0)   # Lasso and Ridge are two methods under Regularization
                                                   # regularization prevents over fitting
    
    
    #Train the model
regression_model.fit(X_train,Y_train) 

    #Predict
Y_predict = regression_model.predict(X_test)
    
    
    #Regression Coefficients
print('Coefficients: ', regression_model.coef_)  
Coefficients = regression_model.coef_

# Regression INtercepts
print("Intercept: ", regression_model.intercept_)
Intercepts = regression_model.intercept_

    
    #Variance : 1 means perfect prediction
print("Variance score i.e Accuracy: {}".format(regression_model.score(X_test,Y_test)*100)) 
#Accuracy is the number of correctly predicted data points out of all the data points.


#RMSE for the model
print("RMSE:",np.sqrt(metrics.mean_squared_error(Y_test, Y_predict)))   #we get rmse=48.01 which is good

# Since you predicted for X_test, Y_test will be the original values

#y_truearray-like of shape (n_samples,) or (n_samples, n_outputs)
#Ground truth (correct) target values.

#y_predarray-like of shape (n_samples,) or (n_samples, n_outputs)
#Estimated target values.


#Plotting for ONe variable at a time
# If you want to plot all the variables simultaneously then plot a 3D graph
plt.scatter(X_test["Open"], Y_test)
plt.plot(X_test["Open"], Coefficients[0] * X_test["Open"] + Intercepts) 
plt.xlabel("Open Price")
plt.ylabel("new Column")
plt.legend()
plt.grid(True)
plt.show()


plt.scatter(X_test["Change %"], Y_test)        #scatter the original test set
plt.plot(X_test["Change %"], Coefficients[3] * X_test["Change %"] + Intercepts)  #plot line for the predicted Y which was 
plt.xlabel("Change %")                                                           #which was predicted for x_test
plt.ylabel("new Column")
plt.legend()
plt.grid(True)
plt.show()








```

```python
# Completing column values for Pred 

for i in range(511,410,-1):
    data["Pred"][i]= ((Coefficients[0]* data["Open"].iloc[i]) +
                         (Coefficients[1] * data["High"].iloc[i])+
                         (Coefficients[2] * data["Low"].iloc[i]) +
                         (Coefficients[3] * data["Change %"].iloc[i]))
    
data["Pred"] = data["Pred"]
```

```python
# CAPM Analysis and Beta Calculation using Regression

import pandas as pd
import numpy as np
import statsmodels.regression.linear_model as lm
import statsmodels.tools.tools as st
import statsmodels.api as sm

#monthly_returns = monthly_prices.pct_change(1)
#returns are just percentage changes i.e Change%


# CAPM model data
data1  = pd.read_csv(r"C:\Users\Suman\Desktop\AICTE Internship\Mid_Cap\SUNTV.csv")
data1['Date'] = pd.to_datetime(data1['Date'], errors='coerce')
data1['Month'] = data1['Date'].dt.strftime('%B') 
data1['Day_Perc_Change_SUN'] = data1['Close Price'].pct_change()
data1 = data1.dropna()



data2 = pd.read_csv(r"C:\Users\Suman\Desktop\AICTE Internship\Nifty50\Nifty50.csv")
data2['Date'] = pd.to_datetime(data2['Date'], errors='coerce')
data2['Month'] = data2['Date'].dt.strftime('%B') 
data2['Day_Perc_Change_Nifty'] = data2['Close'].pct_change()
data2 = data2.dropna()


Change_df= pd.concat([data1["Day_Perc_Change_SUN"],data2["Day_Perc_Change_Nifty"]], axis=1)
Change_df = Change_df.astype(np.float64)

X = Change_df["Day_Perc_Change_SUN"].tail(90)
y = Change_df["Day_Perc_Change_Nifty"].tail(90)


plt.figure(figsize=(20,10))
X.plot()
y.plot()
plt.ylabel("Daily return of SUNTV and Nifty")
plt.show()

# add a constant to the independant variable
X1 = sm.add_constant(X)

# Syntax of OLS
#class statsmodels.regression.linear_model.OLS(endog, exog=None, missing='none', hasconst=None, **kwargs)[source]
# OLS( y:dependant variable, x:independant variable)
capm = sm.OLS(y,X1).fit()


print(capm.summary())

#beta = np.round((np.cov(Change_df["Day_Perc_Change_Nifty"],Change_df["Day_Perc_Change_MF"]))/(np.var(Change_df["Day_Perc_Change_MF"])))

#print("CAPM Beta: ",beta )
```

```python
# Alternatively scipy linear regression

from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X,y)

beta = slope
alpha = intercept
print("Slope: ",slope)
print("Intercept: ",intercept)


```

```python
# Linear Regression Plot

X2 = np.linspace(X.min(),X.max(),100)
Y_hat = X2 * beta + alpha

plt.figure(figsize=(10,7))
plt.scatter(X,y)
plt.xlabel("SUNTV Daily Return for last 3 Months")
plt.ylabel("Nifty Daily Return for last 3 Months")
plt.legend()
plt.plot(X2, Y_hat, "r")

```
