```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split     #when tried without splitting ,overfitting is facedbn
import seaborn as sns
from sklearn import metrics
import sklearn
import math

data = pd.read_csv(r'C:\Users\Suman\Desktop\AICTE Internship\Gold.csv')

data['Change %'] = data['Change %'].str.replace(r'%', '').astype(float)  #removed % from the column values 

#X = pd.concat([data[['Open', 'High','Low','Change %']]])  #Double brackets after data
X = pd.concat([data[['Open', 'High','Low','Change %']].head(400)])
Y = pd.concat([data[['Pred']].head(400)])


#Splitting the data into training and testing set
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size= 0.2, random_state=1)
    
    
    #Create a linear Regression Object
regression_model = LinearRegression()   
    
    
    #Train the model
regression_model.fit(X_train,Y_train) 
    
    
    #Predict
Y_predicted = regression_model.predict(X_test)
    
    
    #Regression Coefficients
print('Coefficients: ', regression_model.coef_)  
Coefficients = regression_model.coef_

    
    #Variance : 1 means perfect prediction
print("Variance score: {}".format(regression_model.score(X_test,Y_test))) 


#RMSE for the model
print("RMSE:",np.sqrt(metrics.mean_squared_error(Y_test, Y_predicted)))

    
#plt.style.use('fivethirtyeight') 
  
## plotting residual errors in training data 
plt.scatter(regression_model.predict(X_train), regression_model.predict(X_train) - Y_train, 
           color = "green", s = 10, label = 'Train data') 
  
# plotting residual errors in test data 
plt.scatter(regression_model.predict(X_test), regression_model.predict(X_test) - Y_test, 
            color = "blue", s = 10, label = 'Test data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 1000 ,linewidth = 2) 


## plotting legend 
plt.legend(loc = 'upper right') 
      
## plot title 
plt.title("Residual errors") 
  
## function to show plot 
plt.show() 
```
